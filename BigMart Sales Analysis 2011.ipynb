```python
# Get Sales for the Year 2011

sales_2011=sales[sales['Year']==2011]
print (" Getting Sales Data for Year 2011")

sales_2011
```

```python
print (" Getting Amount for Each Month")
sales_2011_month = sales_2011.groupby('Month').sum()['Amount']

sales_2011_month
```

```python
import matplotlib.ticker as mtick

plt.figure(figsize=(15, 10))

# Plot the data points
plt.plot(sales_2011_month.index, sales_2011_month.values, marker='o', linestyle='-')

# Add labels to each data point with $ and in thousands
for i, value in enumerate(sales_2011_month.values):
    formatted_value = "{:,.2f}K".format(value / 1000.0)
    plt.text(sales_2011_month.index[i], value, formatted_value, ha='center', va='bottom')

# Format y-axis ticks as currency in thousands
plt.gca().yaxis.set_major_formatter(mtick.StrMethodFormatter('${x:,.0f}K'))

plt.title("Sales Figure of 2011 (per month)")
plt.xlabel('Month')
plt.ylabel('Sales Values')
plt.show()
```

```python
# Plot Total Sales Per Month for Year 2011 as Bar Chart


plt.figure(figsize=(15,10))

plt.bar(sales_month.index,sales_month.values,width=0.8)
plt.plot(sales_month.index,sales_month.values,'r')

plt.xlabel('Month')
plt.ylabel('Sales vales')
plt.title('Sales Per Month in Year 2011')
plt.show()
```

```python
print (" Total Quantity Sold Each Month")


Total_Quantity_sold_by_Coutries = sales_2011.groupby('Month').sum()['Quantity']

Total_Quantity_sold_by_Coutries
```

```python
plt.figure(figsize=(15, 10))


# Plot the data points
plt.plot(Total_Quantity_sold_by_Coutries.index, Total_Quantity_sold_by_Coutries.values, marker='o', linestyle='-')

# Add labels to each data point with $ and in thousands
for i, value in enumerate(Total_Quantity_sold_by_Coutries.values):
    formatted_value = "{:,.2f}K".format(value / 1000.0)
    plt.text(Total_Quantity_sold_by_Coutries.index[i], value, formatted_value, ha='center', va='bottom')

# Format y-axis ticks as currency in thousands
plt.gca().yaxis.set_major_formatter(mtick.StrMethodFormatter('${x:,.0f}K'))

plt.title("Total Quantity Figure of 2011 (Per Month)")
plt.xlabel('Month')
plt.ylabel('Quantity Values')
plt.show()
```

```python
sales_country_wise = sales_2011.groupby('Country').sum()['Amount']
sales_country_wise
```

```python
# show the Top 10 countries based on the highest contributes towards sales.

sales_country_wise = sales_2011.groupby('Country').sum()['Amount']

top_countries = 10

top_countries_data = sales_country_wise.sort_values(ascending=False).head(top_countries)

plt.figure(figsize=(10, 6))

top_countries_data.plot(kind='barh', color='skyblue')

plt.title(f"Top {top_countries} Contributing Countries For 2011")
plt.xlabel("Amount")
plt.ylabel("Country")

plt.show()
```

Plot Scatter Plot for the invoice amounts and see the 10 concentration of the amount by invoice.

```python
# show the Top 10 countries based on the lowest contributes towards sales.

sales_country_wise = sales_2011.groupby('Country').sum()['Amount']

top_countries = 10

top_countries_data = sales_country_wise.sort_values(ascending=True).head(top_countries)

plt.figure(figsize=(10, 6))

top_countries_data.plot(kind='barh', color='skyblue')

plt.title(f"Top {top_countries} Lowest Contributing Countries For 2011")
plt.xlabel("Amount")
plt.ylabel("Country")

plt.show()
```

```python

# Assuming 'Amount' is the column representing sales
sales_invoice_wise = sales_2011.groupby('InvoiceNo').sum()['Amount']



# Find the top 10 invoice numbers with the highest sales
top_10_invoices = sales_invoice_wise.nlargest(10)



# Create a scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(top_10_invoices.index, top_10_invoices.values, color='blue', alpha=0.7)

plt.title("Top 10 Invoice Numbers with the Highest Sales")
plt.xlabel("Invoice Number")
plt.ylabel("Total Sales Amount")
plt.grid(True)
plt.show()


```
